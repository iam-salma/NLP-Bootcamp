```python
import pandas as pd
messages=pd.read_csv('../smsspamcollection/SMSSpamCollection', sep='\t',names=["label","message"])
messages
```

```python
messages.shape
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk import word_tokenize
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    words = word_tokenize(review)
    review = [lemmatizer.lemmatize(word) for word in words if word.isalpha()]
    review = ' '.join(review)
    corpus.append(review)
```

```python
[[i,j,k] for i,j,k in zip(list(map(len,corpus)),corpus, messages['message']) if i<1]
```

```python
corpus
```

```python
from nltk import sent_tokenize, word_tokenize
from gensim.utils import simple_preprocess
```

```python
sentences=[]
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for token in sent_token:
        sentences.append(simple_preprocess(token))
```

```python
def avg_word2vec(sentence):
    # remove out-of-vocabulary words
    #sent = [word for word in doc if word in model.wv.index_to_key]
    #print(sent)
    
    return np.mean([model.wv[word] for word in sentence if word in model.wv.index_to_key],axis=0)
                #or [np.zeros(len(model.wv.index_to_key))], axis=0)
```

```python
from tqdm import tqdm
```

```python
#apply for the entire sentences
import numpy as np
X=[]
for i in tqdm(range(len(sentences))):
    X.append(avg_word2vec(sentences[i]))
```
